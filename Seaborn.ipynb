```python
import collections
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
f = open('C:\\Python37\\Scripts\\BigDataProject\\Cleaned\\a94amj.txt','r')
words = [x for y in [l.split() for l in f.readlines()] for x in y]
data = sorted([(w, words.count(w)) for w in set(words)], key = lambda x:x[1], reverse=True)[:10] 

print(data)

most_words = [x[0] for x in data]
times_used = [int(x[1]) for x in data]
print(most_words)
print(times_used)
```

```python
df = pd.DataFrame({"Count": [113, 102, 72, 56, 51, 49, 47, 46, 39, 38]},
                  index=['human', 'resource', 'systems', 'control', 'management', 'performance', 'commitment', 'turnover', 'the', 'system'])

sns.heatmap(df, annot=True, fmt="g", cmap='viridis')

plt.show()
```
